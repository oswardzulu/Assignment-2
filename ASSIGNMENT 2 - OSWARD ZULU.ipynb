```python
# Imports needed to complete this assignment
import pandas as pd
```

```python
#Read the CSV file into a dataframe
ibm = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
ibm.head()
```

```python
#1. SHOW A BREAKDOWN OF DISTANCE FROM HOME BY JOB ROLE AND ATTRITION.

dist_by_job_and_attrition = ibm.groupby(['DistanceFromHome','Attrition', 'JobRole']).agg({'JobRole': ['count']})
dist_by_job_and_attrition.head(35)
```

```python
#ALSO SHOWING A BREAKDOWN OF DISTANCE FROM HOME BY JOB ROLE.

dist_by_job = ibm.groupby(['DistanceFromHome','JobRole']).agg({'JobRole': ['count']})
dist_by_job.head(20)
```

```python
#ALSO SHOWING A BREAKDOWN OF DISTANCE FROM HOME BY ATTRITION.

dist_by_attrition = ibm.groupby(['DistanceFromHome','Attrition']).agg({'Attrition': ['count']})
dist_by_attrition.head(20)
```

```python
#2. COMPARE THE AVERAGE MONTHLY INCOME BY EDUCATION AND ATTRITION.

avg_month_income_by_edu_and_attrition = ibm.groupby(['EducationField','Attrition']).agg({'MonthlyIncome': ['mean']})
avg_month_income_by_edu_and_attrition
```

```python
#ALSO COMPARING THE AVERAGE MONTHLY INCOME BY EDUCATION.

avg_month_income_by_education = ibm.groupby(['EducationField']).agg({'MonthlyIncome': ['mean']})
avg_month_income_by_education
```

```python
#ALSO COMPARING THE AVERAGE MONTHLY INCOME BY ATTRITION.

avg_month_income_by_attrition = ibm.groupby(['Attrition']).agg({'MonthlyIncome': ['mean']})
avg_month_income_by_attrition
```
